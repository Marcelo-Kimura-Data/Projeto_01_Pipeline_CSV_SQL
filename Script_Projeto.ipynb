## Etapa 2: Tratando os dados

#### Começaremos consolidando as tabelas de vendas

```python
def corrige_titulo(df, ano):
    df = df.copy()
    df.columns = df.columns.str.lower().str.strip().str.replace(" ", "_")
    df["ano_venda"] = ano
    return df
```

```python
df_vendas2020 = corrige_titulo(df_vendas2020, 2020)
df_vendas2021 = corrige_titulo(df_vendas2021, 2021)
df_vendas2022 = corrige_titulo(df_vendas2022, 2022)
```

```python
df_vendas = pd.concat([df_vendas2020, df_vendas2021, df_vendas2022], ignore_index=True)
```

#### Iremos ajustar a tabela de clientes
Precisamos eliminar as colunas 10 e 11 e além disso limpar as linhas com espaços em branco.\
OBS: tipo de dado de documento precisa ser string para não se perder informação.

```python
# df_clientes = df_clientes.drop(columns=['Unnamed: 10','Unnamed: 11'],errors='ignore')
# A coluna Unnamed 10 não estava aparecendo
# nesse caso, iremos repetir o processo de importação da base de dados aplicando mais dois argumentos no read_excel para ajustar as colunas
```

```python
clientes = 'C:/Users/Kimura/Desktop/Projetos/Projeto_01/Dados/Cadastro Clientes.xlsx'
df_clientes = pd.read_excel(clientes, header=2, dtype={'Documento': str})
df_clientes.head(10)
df_clientes = df_clientes.replace(r'^\s*$', pd.NA, regex=True)
df_clientes = df_clientes.dropna(axis=1, how='all')
df_clientes = df_clientes.dropna(axis=0, how='all')
df_clientes.info()
df_clientes.head(10)
```

```python
df_clientes['Data Nascimento'] = pd.to_datetime(df_clientes['Data Nascimento'], errors='coerce')
```

```python
def padroniza_colunas(df):
    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(' ', '_')
        .str.replace(r'[^\w]', '', regex=True)
    )
    return df
```

```python
df_clientes = padroniza_colunas(df_clientes)
df_clientes.head(10)
```

```python
df_clientes['genero'] = df_clientes['genero'].replace({'M':'Masculino', 'F':'Feminino'})
df_clientes['estado_civil'] = df_clientes['estado_civil'].replace({'C':'Casado','S':'Solteiro'})
df_clientes['primeiro_nome'] = df_clientes['primeiro_nome'].str.title() + ' ' + df_clientes['sobrenome'].str.title()
df_clientes = df_clientes.drop(columns=['sobrenome'])
df_clientes = df_clientes.rename(columns={'primeiro_nome': 'nome_completo'})
```

```python
df_clientes.head(10)
```

#### Tratando dados de Lojas e Localidades

```python
df_localidades = padroniza_colunas(df_localidades)
lojas = 'C:/Users/Kimura/Desktop/Projetos/Projeto_01/Dados/Cadastro Lojas.xlsx'
df_lojas = pd.read_excel(lojas, dtype={'Documento Gerente': str})
df_lojas = padroniza_colunas(df_lojas)
```
