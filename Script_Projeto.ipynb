### 3.3: Iremos ajustar a tabela de clientes
- Para ajustar as 2 primeiras linhas erradas iremos basicamente importar novamente o dataframe adicionando a informação `header=2`;
- Uma observação para a coluna 'Documento' que precisa ser do tipo `str`;
- Em seguida para ajustar as 2 colunas com valores nulos, iremos substituir os valores nulos por `NA` e depois aplicar um código para excluir tudo;
- Iremos usar novamente a função para ajustar os nomes das colunas;
- Iremos ajustar o tipo de dado da coluna Data Nascimento;
- Ajustar os valores de Genero e Estado Civil;
- Concatenar o Primeiro Nome com Sobrenome.

```python
# Importação do df já atribuindo o título correto, eliminando as 2 linhas em branco e ajustando o tipo de dado da coluna documento:
clientes = 'C:/Users/Kimura/Desktop/Projetos/Projeto_01/Dados/Cadastro Clientes.xlsx'
df_clientes = pd.read_excel(clientes, header=2, dtype={'Documento': str})

# Transformando valores em branco em NA e droppando do DF (linhas e colunas):
df_clientes = df_clientes.replace(r'^\s*$', pd.NA, regex=True)
df_clientes = df_clientes.dropna(axis=1, how='all')
df_clientes = df_clientes.dropna(axis=0, how='all')

# Ajustando os nomes das colunas:
df_clientes = padroniza_colunas(df_clientes)

# Ajustando o tipo de dado da coluna de 'data_nascimento':
df_clientes['data_nascimento'] = pd.to_datetime(df_clientes['data_nascimento'], errors='coerce')

# Ajustando os valores da coluna 'genero' e 'estado_civil' (escrevendo por extenso):
df_clientes['genero'] = df_clientes['genero'].replace({'M':'Masculino', 'F':'Feminino'})
df_clientes['estado_civil'] = df_clientes['estado_civil'].replace({'C':'Casado','S':'Solteiro'})

# Concatenando o 'primeiro_nome' com 'sobrenome' em uma nova coluna 'nome_completo':
df_clientes['primeiro_nome'] = df_clientes['primeiro_nome'].str.title() + ' ' + df_clientes['sobrenome'].str.title()
df_clientes = df_clientes.drop(columns=['sobrenome'])
df_clientes = df_clientes.rename(columns={'primeiro_nome': 'nome_completo'})
```

### 3.4: Tratando dados de Lojas e Localidades

- Iremos ajustar o df_loja por conta da coluna Documento Gerente;
- Em seguida iremos criar um novo df com lojas e localidades;
- Ajuste no nome do gerente.

```python
# Importando os dados já com o ajuste do tipo de dado do Documento Gerente e ajuste nos nomes das colunas:
lojas = 'C:/Users/Kimura/Desktop/Projetos/Projeto_01/Dados/Cadastro Lojas.xlsx'
df_lojas = pd.read_excel(lojas, dtype={'Documento Gerente': str})
df_lojas = padroniza_colunas(df_lojas)

# Criação do df onde iremos juntar todas as colunas e ajuste no nome do gerente:
df_lojas_full = pd.merge(df_lojas, df_localidades, on='id_localidade', how='left')
df_lojas_full['gerente_loja'] = df_lojas_full['gerente_loja'].apply(lambda x: ' '.join(x.split(', ')[::-1]))
```

### 3.5: Ajustando df de produtos
- Iremos eliminar a coluna 'obs';
- Separar as informações de produto e cor da coluna 'produto';
- df de devoluções está em conformidade.

```python
# Eliminando coluna 'obs':
df_produtos = df_produtos.drop(columns=['obs'])

# Separando nome produto de cor de produto:
df_produtos[['produto', 'cor']] = df_produtos['produto'].str.rsplit(' - ', n=1, expand=True)
# .rsplit(' - ', n=1, expand=True) faz -> divide a coluna a partir do ' - ';
# n=1 significa divida no máximo 1 vez
# expand=True -> o resultado vira um df com 2 colunas.
```
